# import poverty and medain household income historical data

```python
year = house_data["year"].unique().tolist()
```

```python
import requests
import json
from pprint import pprint
import time
```

```python
counties_poverty_df_list = []
for i in range(1,len(year)-2):
    # URL for county poverty data
    poverty_url = f'https://api.census.gov/data/timeseries/poverty/saipe?get=NAME,SAEMHI_PT,SAEMHI_MOE,SAEPOVALL_PT,SAEPOVALL_MOE,SAEPOVRTALL_MOE,SAEPOVRTALL_PT,STABREV&for=county:*&in=state:34&YEAR={year[i]}'
    # API call to get countries metadata
    poverty_data = requests.get(poverty_url).json()
    time.sleep(1)
    # initialize empty lists
    county_name = []
    median_hh_income = []
    median_hh_inc_moe = []
    poverty_count = []
    poverty_count_moe = []
    poverty_rate = []
    poverty_rate_moe = []
    st_abb =[]
    state_fips = []
    county_fips = []
    year_rec=[]
    # iterate through the response to populate the lists and create a dataframe
    for i in range(1,len(poverty_data)):
        county_name.append(poverty_data[i][0])
        median_hh_income.append(poverty_data[i][1])
        median_hh_inc_moe.append(poverty_data[i][2])
        poverty_count.append(poverty_data[i][3])
        poverty_count_moe.append(poverty_data[i][4])
        poverty_rate.append(poverty_data[i][5])
        poverty_rate_moe.append(poverty_data[i][6])
        st_abb.append(poverty_data[i][7])
        year_rec.append(poverty_data[i][8])
        state_fips.append(poverty_data[i][9])
        county_fips.append(poverty_data[i][10])
    counties_poverty_df = pd.DataFrame({"county": county_name,
                                        "median_hh_income": median_hh_income,
                                        "median_hh_inc_moe": median_hh_inc_moe,
                                        "poverty_count": poverty_count,
                                        "poverty_count_moe": poverty_count_moe,
                                        "poverty_rate": poverty_rate,
                                        "poverty_rate_moe": poverty_rate_moe,
                                        "st_abb": st_abb,
                                        'year':year_rec,
                                        "state_fips": state_fips,
                                        "county_fips": county_fips})
    counties_poverty_df_list.append(counties_poverty_df)
pov_mhi_df = pd.concat(counties_poverty_df_list, ignore_index = True)
```

```python
pov_mhi_df.head(-10)
```

```python
pov_mhi_df['county']= pov_mhi_df.county.str.replace("County","")
pov_mhi_df['county'] = pov_mhi_df['county'].str.strip()
pov_mhi_df['year'] = pov_mhi_df['year'].astype(int)
pov_mhi_df.head(-10)
```

```python
pov_mhi_df.to_csv('pov_mhi.csv', index = False)
```

# import APR dataset

```python
df = pd.read_csv('apr_history.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Week'] = pd.to_datetime(df['Week'], format='%m/%d/%Y')
df['year'] = pd.DatetimeIndex(df['Week']).year
df.head()
```

```python
apr_data = df.groupby(['year'], as_index = False).agg(frm_30 = ('FRM_30', 'mean'), points_30 = ('points_30', 'mean')\
                                                     , frm_15 = ('FRM_15', 'mean'), points_15 = ('points_15', 'mean'))
apr_data.head()
```

```python
apr_data.dropna(how='any',inplace = True)
apr_data.reset_index(drop=True, inplace =True)
apr_data.head()
```
